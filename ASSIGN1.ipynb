```python
#TASK 6
# Program to add two numbers with user input

# Take the first number as input from the user
# input() returns a string, so we convert it to a float to handle decimals
num1 = float(input("Enter first number: "))

# Take the second number as input from the user
num2 = float(input("Enter second number: "))

# Add the two numbers
sum_result = num1 + num2

# Display the sum using an f-string for clear formatting
print(f"The sum of {num1} and {num2} is {sum_result}")
```

```python
#TASK 7
# Program to calculate total marks and percentage for five subjects

print("Enter the marks obtained in 5 subjects (out of 100 each):")

# Take input for each subject and convert to float for potentially non-integer marks
sub1 = float(input("Subject 1 marks: "))
sub2 = float(input("Subject 2 marks: "))
sub3 = float(input("Subject 3 marks: "))
sub4 = float(input("Subject 4 marks: "))
sub5 = float(input("Subject 5 marks: "))

# Calculate the total marks obtained
total_marks_obtained = sub1 + sub2 + sub3 + sub4 + sub5

# Assume maximum marks per subject is 100
max_marks_per_subject = 100
number_of_subjects = 5
total_max_marks = max_marks_per_subject * number_of_subjects # which is 500

# Calculate the percentage
# Formula: (Total Marks Obtained / Total Maximum Marks) * 100
percentage = (total_marks_obtained / total_max_marks) * 100

# Print the results, formatting the percentage to two decimal places
print("-" * 30)
print(f"Total Marks Obtained: {total_marks_obtained} out of {total_max_marks}")
print(f"Percentage: {percentage:.2f}%")
print("-" * 30)
```

```python
#TASK 8
# Program to check if a single number is even or odd

# Take input from the user and convert it to an integer
try:
    num = int(input("Enter a number: "))

    # Check if the number is even using the modulo operator
    # The modulo operator (%) returns the remainder of a division.
    # If the remainder when divided by 2 is 0, the number is even.
    if num % 2 == 0:
        print(f"The number {num} is Even.")
    else:
        print(f"The number {num} is Odd.")

except ValueError:
    print("Invalid input. Please enter a valid integer.")
```
